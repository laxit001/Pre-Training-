```python
from datasets import load_dataset
import os
```

```python
dataset=load_dataset("text",data_files="/content/data_set.txt")
```

```python
dataset
```

```python
from transformers  import (
    GPT2TokenizerFast,
    GPT2Config,
    GPT2LMHeadModel,
    DataCollatorForLanguageModeling,
    Trainer,
    TrainingArguments
)
```

```python
tokenizer=GPT2TokenizerFast.from_pretrained("gpt2")
tokenizer.pad_token=tokenizer.eos_token
```

```python
def tokenize_function(examples):
  return tokenizer(examples["text"])
```

```python
tokenized_dataset=dataset.map(tokenize_function,batched=True,num_proc=4,remove_columns=["text"])
```

```python
tokenized_dataset
```

```python
config=GPT2Config(
    vocab_size=tokenizer.vocab_size,
    bos_token_id=tokenizer.bos_token_id,
    eos_token_id=tokenizer.eos_token_id,
    n_layer=6,
    n_head=6,
    n_embd=384)

model=GPT2LMHeadModel(config)
```

```python
data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer,mlm=False)
```

```python
training_args=TrainingArguments(
    output_dir='./gpt2-generic_model',
    eval_steps=500,
    logging_steps=100,
    save_steps=1000,
    save_total_limit=2,
    per_device_eval_batch_size=4,
    num_train_epochs=3,

)
```

```python
trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    data_collator=data_collator,
    tokenizer=tokenizer
)
```

```python
print(f"Number of samples in training dataset: {len(tokenized_dataset['train'])}")
print(f"Training batch size: {training_args.per_device_train_batch_size}")
```

```python
trainer.train()
```

```python
model.save_pretrained("/content/gpt2-generic_model")
```

```python
from transformers import pipeline

generator=pipeline('text-generation',model='/content/gpt2-generic_model',tokenizer='gpt2')

prompt="what is cricket"
output=generator(prompt,max_length=100,num_return_sequences=1)
print(output[0]['generated_text'])
```
